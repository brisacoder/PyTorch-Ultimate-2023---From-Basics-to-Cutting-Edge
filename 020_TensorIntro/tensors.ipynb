```python
import torch
```

```python

x11 = torch.tensor(2.0, requires_grad=True)
x21 = torch.tensor(3.0, requires_grad=True)

x12 = 5 * x11 - 3 * x21
x22 = 2 * x11**2 + 2 * x21

y = 4 * x12 + 3 * x22
z = x12 * x22

dy_dx11 = torch.autograd.grad(y, x11, retain_graph=True)[0]
dz_dx11 = torch.autograd.grad(z, x11)[0]

print("∂y/∂x11 =", dy_dx11)
print("∂z/∂x11 =", dz_dx11)
```

```python
x11 = torch.tensor(2.0, requires_grad=True)
x21 = torch.tensor(3.0, requires_grad=True)

x12 = 5 * x11 - 3 * x21
x22 = 2 * x11**2 + 2 * x21

y = 4 * x12 + 3 * x22
z = x12 * x22

y.backward()
# this will error if x11 is not set to require_grad=True
try:
    z.backward()
except RuntimeError as e:
    print("Error:", e)

print("∂y/∂x11 =", x11.grad)
print("∂z/∂x11 =", x11.grad)
```

```python
x11 = torch.tensor(2.0, requires_grad=True)
x21 = torch.tensor(3.0, requires_grad=True)

x12 = 5 * x11 - 3 * x21
x22 = 2 * x11**2 + 2 * x21

y = 4 * x12 + 3 * x22
z = x12 * x22

y.backward(retain_graph=True)
print("∂y/∂x11 =", x11.grad)
x11.grad.zero_()  # Reset gradients to avoid accumulation
try:
    z.backward()
except RuntimeError as e:
    # Exception unless retain_graph=True
    print("Error:", e)

# wrong values for both since graph computation was retained and z.backward added to
# it
print("∂z/∂x11 =", x11.grad)
```

```python
import torch

x = torch.tensor(2.0, requires_grad=True)
y1 = x**2          # dy1/dx = 2x = 4
y2 = 3 * x         # dy2/dx = 3

# First call: retain the graph
dy1_dx = torch.autograd.grad(
    [y1, y2], x,
    grad_outputs=[torch.tensor(1.0), torch.tensor(0.0)],
    retain_graph=True
)[0]

# Second call: reuse the graph
dy2_dx = torch.autograd.grad(
    [y1, y2], x,
    grad_outputs=[torch.tensor(0.0), torch.tensor(1.0)]
)[0]

print("∂y1/∂x =", dy1_dx)  # 4.0
print("∂y2/∂x =", dy2_dx)  # 3.0
```
