```python
# Functions
def relu(x):
    return np.maximum(x, 0, x)

def hidden_nodes(X):
    G = np.dot(X, input_weights)
    G = G + biases
    H = relu(G)
    return H

def predict(X):
    out = hidden_nodes(X)
    out = np.dot(out, beta)
    return out
```
