```python
import numpy as np
from sklearn.metrics import pairwise_distances

X = np.array([0, 1])
w1 = np.array([2, 3])
w2 = np.array([0.4, 1.8])
```

```python
pairwise_distances(w1.reshape(1, -1), w2.reshape(1, -1), metric='cosine')
```

```python
w1_norm = np.linalg.norm(w1)
w2_norm = np.linalg.norm(w2)
w1_minus_w2_norm = np.linalg.norm(w1-w2)
```

```python
# Calculate the cosine of the angle between w1 and w2 using the law of cosines

cos_a = (w1_norm ** 2 + w2_norm ** 2 - w1_minus_w2_norm ** 2)/(2 * w1_norm * w2_norm)
cos_a
```

```python
# Cosine distance
1 - cos_a
```
